```python
from music21 import converter

# Load your MusicXML file
score = converter.parse(r"c:/xampp/htdocs/Mobile-Charter/Charting-Device/test_output.musicxml")

# Detect key
key = score.analyze('key')
print(f"Detected key: {key.tonic.name} {key.mode}")

# Movable-do solfège mapping
solfege_map = {
    1: "do",
    2: "re",
    3: "mi",
    4: "fa",
    5: "so",
    6: "la",
    7: "ti"
}

# Loop through notes and print solfège
for note in score.recurse().notes:
    degree = key.getScaleDegreeFromPitch(note.pitch)
    syllable = solfege_map.get(degree, "?")
    print(f"{note.nameWithOctave} → {syllable}")
```
